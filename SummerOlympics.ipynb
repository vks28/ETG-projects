```python
import pandas as pd
```

```python
df=pd.read_csv("summer.csv")
```

1. In how many cities Summer Olympics is held so far?

```python
df["City"].nunique()
```

```python
list(df["City"].unique())
```

2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
print(df[df["Medal"]=="Gold"].Sport.value_counts().head())
df[df["Medal"]=="Gold"].Sport.value_counts().head().plot.bar(figsize = (10,10))
```

3. Which sport is having most number of medals so far? (Top 5)

```python
print(df.Sport.value_counts().head())
df.Sport.value_counts().head().plot.line(figsize = (10,10))
```

4. Which player has won most number of medals? (Top 5)

```python
print(df.Athlete.value_counts().head())
df.Athlete.value_counts().head().plot.bar(figsize = (10,10))
```

5. Which player has won most number Gold Medals of medals? (Top 5)

```python
print(df[df.Medal=="Gold"].Athlete.value_counts().head())
df[df.Medal=="Gold"].Athlete.value_counts().head().plot.line(figsize = (10,10))
```

6. In which year India won first Gold Medal in Summer Olympics?

```python
dfGold=df[df.Medal=="Gold"]
min(dfGold[dfGold.Country=="IND"].Year)
```

7. Which event is most popular in terms on number of players? (Top 5)

```python
print(df["Event"].value_counts().head())
df["Event"].value_counts().head().plot.bar(figsize = (10,10))
```

8. Which sport is having most female Gold Medalists? (Top 5)

```python
dfGold=df[df.Medal=="Gold"]
print(dfGold[dfGold.Gender=="Women"].Sport.value_counts().head())
dfGold[dfGold.Gender=="Women"].Sport.value_counts().head().plot.bar(figsize = (10,10))
```
